```python
#use dictionary to created the table for hospitals performing the services.
data = {'MEDICAL SERVICES':['Public_Sector','Emergency_Transport','Maternal_health_delivery_services','c_section_yn','PHCN_Electricity','Child_Healthcare','Malaria','Antenatal','VACCINES_refridgerator','Improved_Sanitation','Improved_water supply','Family_Planning','Skilled_Birth_Attendant'],
        'SERVICES':[24763,10768,20920,5246,11974,21504,26143,24441,6270,15481,13931,19415,13455]}

SERVICES = DataFrame(data)


SERVICES
```

```python
#use dictionary to created the table for hospitals not performing the services.
data2 = {'MEDICAL SERVICES':['Public_Sector','Emergency_Transport','Maternal_health_delivery_services','c_section_yn','PHCN_Electricity','Child_Healthcare','Malaria','Antenatal','VACCINES_refridgerator','Improved_Sanitation','Improved_water supply','Family_Planning','Skilled_Birth_Attendant'],
        'NO_SERVICES':[9376,23371,13219,28893,22165,12635,7996,9698,27869,18658,20208,14724,20694]}

NO_SERVICES = DataFrame(data2)


NO_SERVICES
```

```python
#use pd.merge(SERVICES,NO_SERVICES,how='inner',on='HOSPITALS') to merge data and data2 created above
df4= pd.merge(SERVICES,NO_SERVICES,how='inner',on='MEDICAL SERVICES')
df4
```
